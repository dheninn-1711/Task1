```python
# Import the pandas library
import pandas as pd
```

```python
# Load the dataset from a CSV file into a DataFrame
anime_df = pd.read_csv("C:/Users/dheni/Downloads/animes.csv")
anime_df.head(5)
```

```python
# Initial information about the dataset
print("Intial Dataset Info:")
anime_df.info()
```

```python
# Drop rows with missing values
anime_df.dropna(subset=['ranked', 'score'], inplace=True)
```

```python
# Fill missing values
anime_df['episodes'].fillna(0, inplace=True)
anime_df['synopsis'].fillna('No description available', inplace=True)
anime_df.shape #(number of rows and columns)
```

```python
# Remove duplicate rows
anime_df.drop_duplicates(subset=['title','uid'], inplace=True)
anime_df.shape
```

```python
# Convert the column to numeric, coercing errors to NaN
anime_df['score'] = pd.to_numeric(anime_df['score'], errors='coerce') 
anime_df['score'] = (anime_df['score'] * 2).round() / 2 # Round the 'score' values to the nearest 0.5
anime_df
```

```python
# Convert the columns to numeric, coercing errors to NaN, fill NaN with 0, and convert float to integer type
anime_df['ranked'] = pd.to_numeric(anime_df['ranked'], errors='coerce').fillna(0).astype(int)
anime_df['episodes'] = pd.to_numeric(anime_df['episodes'], errors='coerce').fillna(0).astype(int)
anime_df.head(5)
```

```python
# Convert the 'aired' column to datetime format
anime_df['aired'] = anime_df['aired'].str.split(' to ').str[0]  # Get the start date
anime_df['aired'] = pd.to_datetime(anime_df['aired'], format='%b %d, %Y', errors='coerce')  # Convert to datetime

# Change the format of 'aired' to dd-mm-yyyy
anime_df['aired'] = anime_df['aired'].dt.strftime('%d-%m-%Y')
anime_df.head(5)
```

```python
# Count the number of null and non-null values
anime_df['aired'].isnull().value_counts()
```

```python
# Interpolate missing values in the 'aired' column using time-based interpolation
anime_df['aired'] = anime_df['aired'].interpolate(method='time')
anime_df['aired'] = anime_df['aired'].fillna(method='ffill')# Forward fill any remaining NaN values in the 'aired' column
anime_df['aired'].isnull().value_counts()
```

```python
# Clean the 'genre' column
anime_df['genre'] = anime_df['genre'].str.strip("[]").str.replace("'", "").str.split(', ').apply(lambda x: [g.lower().replace(' ', '_') for g in x])
anime_df.head(5)
```

```python
# Rename the 'uid' and 'aired' columns
anime_df.rename(columns={'uid': 'anime_id', 'aired': 'released_date'}, inplace=True)
```

```python
# Convert all column names to uppercase
anime_df.columns = anime_df.columns.str.upper()
anime_df
```

```python
# Cleaned dataset information
print("Cleaned Dataset Info:")
anime_df.info()
```

```python
# Save the cleaned DataFrame to a CSV file
cleaned_file_path = "C:/Users/dheni/Downloads/cleaned_anime_data.csv"  # Specify the path where you want to save the file
anime_df.to_csv(cleaned_file_path, index=False)  # Save the DataFrame without the index
print(f"Cleaned dataset saved to: {cleaned_file_path}")
```
